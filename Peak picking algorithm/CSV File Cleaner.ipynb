```python
import os
import pandas as pd

# Define the folder paths
input_folder_path = r'C:\Users\kulma\OneDrive\Documents\School\MASc Year 1\Surface Profilometry Analyzer\Surface-Profilometry-Analyzer\Peak picking algorithm\Normalized CSV Files'  # Replace with the actual path to your input folder
output_folder_path = 'path_to_your_output_folder'  # Replace with the actual path to your output folder

# Create the output folder if it doesn't exist
os.makedirs(output_folder_path, exist_ok=True)

# Define the headers based on the observed structure
headers = ["V1", "µA1", "V2", "µA2", "V3", "µA3", "V4", "µA4", "V5", "µA5", "V6", "µA6", "V7", "µA7", "V8", "µA8"]

# Iterate over all files in the input folder
for filename in os.listdir(input_folder_path):
    if filename.endswith('.csv'):
        input_file_path = os.path.join(input_folder_path, filename)

        try:
            # Read the CSV file with UTF-16 encoding, skipping the initial rows to capture correct headers
            df = pd.read_csv(input_file_path, encoding='utf-16', skiprows=12, names=headers)

            # Filter columns that contain 'µA' in their header
            ua_columns = [col for col in df.columns if 'µA' in col]
            df_ua = df[ua_columns]

            # Save the filtered DataFrame to a new CSV file in the output folder
            output_file_path = os.path.join(output_folder_path, filename.replace('.csv', '_uA.csv'))
            df_ua.to_csv(output_file_path, index=False)

            print(f"Filtered data saved to {output_file_path}")

            # Save each individual 'µA' column to its own CSV file in the output folder
            for col in ua_columns:
                individual_output_file_path = os.path.join(output_folder_path, filename.replace('.csv', f'_{col}.csv'))
                df[[col]].to_csv(individual_output_file_path, index=False)
                print(f"Column {col} saved to {individual_output_file_path}")

        except Exception as e:
            print(f"Failed to process {filename}: {e}")

print("Processing completed.")
```
