```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Function to find the lowest point using a 3-point peak picking process
def find_lowest_point(data):
    n = len(data)
    lowest_value = data[0]  # Initialize with the first value
    lowest_index = 0
    
    # Iterate through the data to find the lowest point
    for i in range(1, n-1):
        if data[i] < data[i-1] and data[i] < data[i+1]:
            if data[i] < lowest_value:
                lowest_value = data[i]
                lowest_index = i
    
    return lowest_index, lowest_value

# Function to interpolate the baseline
def interpolate_baseline(data):
    if len(data) < 40:
        raise ValueError("Data length is too short to calculate baseline using first and last 20 points.")
    
    left_segment = data[:20]
    right_segment = data[-20:]
    
    # Debug: Print segments
    print("Left segment:", left_segment)
    print("Right segment:", right_segment)
    
    left_segment_mean = np.mean(left_segment)
    right_segment_mean = np.mean(right_segment)
    
    # Debug: Print means
    print("Left segment mean:", left_segment_mean)
    print("Right segment mean:", right_segment_mean)
    
    baseline_slope = (right_segment_mean - left_segment_mean) / (len(data) - 1)
    baseline_intercept = left_segment_mean
    
    # Debug: Print slope and intercept
    print("Baseline slope:", baseline_slope)
    print("Baseline intercept:", baseline_intercept)
    
    baseline = baseline_intercept + baseline_slope * np.arange(len(data))
    
    return baseline

# Load the data from CSV file
file_path = 'C:/Users/kulma/OneDrive/Documents/School/MASc Year 1/Surface Profilometry Analyzer/Surface-Profilometry-Analyzer/Peak picking algorithm/extracted_csv_files/cleaned_20mV 20Hz.csv'
df = pd.read_csv(file_path)

# Check the structure of the data
print(df.head())

# Loop through each second column (uA columns)
for i in range(1, 14, 2):
    uA = df.iloc[:, i]  # Current column
    
    data = uA.to_numpy()

    # Debug: Print first few data points
    print(f"Data for Group {i//2 + 1}:", data[:10])
    
    # Find the lowest point
    lowest_index, lowest_value = find_lowest_point(data)
    
    # Debug: Print lowest point details
    print(f"Lowest point for Group {i//2 + 1}: Index = {lowest_index}, Value = {lowest_value}")

    # Interpolate the baseline
    try:
        baseline = interpolate_baseline(data)
        
        # Debug: Print first few baseline points
        print(f"Baseline for Group {i//2 + 1}:", baseline[:10])

        # Compute the normalized peak value
        normalized_peak = lowest_value - baseline[lowest_index]

        # Debug: Print normalized peak value
        print(f"Normalized peak for Group {i//2 + 1}: {normalized_peak}")

        # Plotting the data with the interpolated baseline
        plt.figure(figsize=(10, 6))
        plt.plot(data, marker='o', linestyle='-', color='b', label=f'Data {i//2 + 1}')
        plt.plot(baseline, linestyle='--', color='r', label='Baseline')
        plt.scatter(lowest_index, lowest_value, color='g', s=100, label='Lowest Point')
        plt.vlines(lowest_index, lowest_value, baseline[lowest_index], color='r', linestyle='--')

        # Annotate the normalized peak value
        plt.annotate(f'Normalized Peak: {normalized_peak:.3f}', 
                     xy=(lowest_index, lowest_value), 
                     xytext=(lowest_index, lowest_value + 0.005), 
                     arrowprops=dict(facecolor='black', shrink=0.05),
                     fontsize=10, color='black')

        plt.title(f'Normalized Peak Relative to Interpolated Baseline (Group {i//2 + 1})')
        plt.xlabel('Index')
        plt.ylabel('Current (uA)')
        plt.legend()

        plt.grid(True)
        plt.tight_layout()
        plt.show()

        print(f"Group {i//2 + 1}")
        print(f"Lowest value found: {lowest_value}")
        print(f"Normalized value of the lowest peak: {normalized_peak}")
        print("\n")
    
    except ValueError as e:
        print(f"Error processing Group {i//2 + 1}: {e}")
```
